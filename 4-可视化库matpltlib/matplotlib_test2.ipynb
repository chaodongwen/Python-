```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel("2018-2019（2）研究生课程安排.xls")
#print(data)
mpl.rcParams['font.sans-serif'] = ['simHei']
mpl.rcParams['axes.unicode_minus'] = False
plt.plot(data["学分"], data["学时"], 'r', lw=2)
plt.xlabel("学分")
plt.ylabel("学时")
plt.title("学分与学时的关系")
```

```python
a = data["选课人数"] 
b = data["学分"]
plt.plot(a, b, 'g', lw=2)
plt.xlabel("选课人数")
plt.ylabel("学分")
plt.title("选课人数与学分的关系")
#plt.scatter(x,y,c=color,s=scale,label=color,alpha=0.6,edgecolors='white')
#plt.scatter(a,b)
```
