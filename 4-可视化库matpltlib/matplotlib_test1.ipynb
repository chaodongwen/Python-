```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
import matplotlib.pylab as plt2
import numpy as np
```

```python
x = np.linspace(-1, 1, 100)
y = x**2
```

```python
mpl.rcParams['font.sans-serif'] = ['simHei']
mpl.rcParams['axes.unicode_minus'] = False
plt.plot(x, y, 'r-', lw=3)
plt2.plot(x, y, 'r-', lw=3)
plt.xlabel("x的值")
plt.ylabel("y的值")
plt.title("二次函数")
plt.show()
```

```python
data = np.random.randn(1000)
mpl.rcParams['font.sans-serif'] = ['simHei']
mpl.rcParams['axes.unicode_minus'] = False
plt.hist(data, bins=20, normed=0, facecolor='green', edgecolor='black')
plt.xlabel("区间")
plt.ylabel("频数/频率")
plt.title("直方图")
plt.show()
```
